```python
import requests
```

```python
api_key='88c70ab0e8c084a2079b36265fdce2b8'
city='London'
url=f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}"

response=requests.get(url)
data=response.json()
```

```python
data
```

```python
temperature = data['main']['temp']
```

```python
description = data['weather'][0]['description']
```

```python
print(f"The temperature in {city} is {temperature} Kelvin and the weather is described as {description}.")
```

```python
temprature_celsius = temperature - 273.15
```

```python
print(f"The temperature in {city} is {temprature_celsius:.2f} °C and the weather is described as {description}.")
```
